- show in the line plot the population of each country in the 62 years 

```python
unique_countries = data['Country Name'].unique()
years = data.columns[1:].astype(int)

fig = go.Figure()

for country_name in unique_countries:
    country_data = data[data['Country Name'] == country_name]
    rows_data = country_data.iloc[:, 1:]

    for i in range(rows_data.shape[0]):
        row_data = rows_data.iloc[i, :]
        fig.add_trace(go.Scatter(x=years, y=row_data, mode='lines+markers', name=country_name))

fig.update_layout(
    title='Population Over the Years for All Countries',
    xaxis=dict(title='Year'),
    yaxis=dict(title='Population'),
    legend=dict(x=1.05, y=2, traceorder='normal', orientation='h'),
    margin=dict(l=0, r=0, t=50, b=0),
)

fig.show()
```

- At the above plot we can show the population in the world is increasing per year 
- we can show the population in the low&middel country and the IDA have the biggest population compare with the another countries 
- we can show the average of the population in all countries are increasing not decreasing that is mean the the birth rate is bigger than the death rate in the all countries from the 1960 to 2022 

```python
years = data.loc[:, '1960':'1969']

# Reshape the DataFrame for Plotly Express
df = pd.melt(data, value_vars=years.columns, var_name='Year', value_name='Population')

fig = px.histogram(df, x='Population', nbins=15, histnorm='probability density', color='Year',
                   title='Population Distribution from 1960 to 1969',
                   labels={'Population': 'Population', 'Year': 'Year'},
                   marginal='rug')

fig.update_layout(
    xaxis=dict(title='Population'),
    yaxis=dict(title='Density'),
    legend=dict(x=1.05, y=1, traceorder='normal', orientation='v'),
    margin=dict(l=0, r=0, t=50, b=0),
)

fig.show()
```

- most populations in the countries are between the 0 to 0.5B in the 10 years from 1960 to 1969
- we have small number of populations in some countries between 3.5B to 4B
- we have small number of populations in some countries between 0.5B to 2B

```python
# Extracting the required data
data_pop = data.loc[:, '1960':'1969']
data_pop['Country Name'] = data.loc[:, 'Country Name']
data_pop['pop_10years'] = data.loc[:, '1960':'1969'].sum(axis=1)
data_pop_sorted = data_pop.sort_values(by='pop_10years', ascending=False).head(50)

# Create horizontal bar chart with Plotly Express
fig = px.bar(data_pop_sorted, x='pop_10years', y='Country Name', color='pop_10years',
             orientation='h', text='pop_10years',
             labels={'pop_10years': 'Population in 10 years', 'Country Name': 'Country Name'},
             title='From 1960 to 1969 population',
             color_continuous_scale='viridis')

fig.update_layout(
    xaxis=dict(title='Population in 10 years'),
    yaxis=dict(title='Country Name'),
    legend=dict(x=1.1, y=1, traceorder='normal', orientation='v'),
    margin=dict(l=0, r=0, t=50, b=0),
)

fig.show()
```

- We can show the higher 50 countries in the population numbers in the dataset in the 10 years from 1960 to 1969
- the number of the population in the world at the 1960 to 1969 greater than  30 B
- the number of the population in Japan at the 1960 to 1969 less than  5 B
